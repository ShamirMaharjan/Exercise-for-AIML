```python
missing_info = df.isna().sum()
missing_percent = (missing_info / len(df)) * 100

# Combine into a DataFrame
missing_df = pd.DataFrame({
    'Missing Count': missing_info,
    'Missing %': missing_percent.round(2)
})

print(missing_df)
```

```python
df["Age"] = df["Age"].fillna(df["Age"].median()) #Used Median cuz its Better for skewed data or data with outliers (more robust), 
#biological measurements often have outliers
print(df["Age"].isna().sum())
df["Age"]
```

```python
df["Height"] = df["Height"].fillna(df["Height"].median()) #Used Median cuz its Better for skewed data or data with outliers (more robust), 
#biological measurements often have outliers
df["Height"]
```

```python
df["Weight"] =df["Weight"].fillna(df["Weight"].median())
df["Weight"]
```

```python
df["BMI"] = df["BMI"].fillna(df["BMI"].median())
df["BMI"]
```

```python
# For Temperature (continuous variable) - Use mean
df['Temperature'] = df['Temperature'].fillna(df['Temperature'].mean())

# For Heart Rate (continuous variable) - Use median
df['Heart Rate'] = df['Heart Rate'].fillna(df['Heart Rate'].median())

# For Cholesterol (continuous variable) - Use median
df['Cholesterol'] = df['Cholesterol'].fillna(df['Cholesterol'].median())
```

```python
# For Gender (categorical) - Use mode (most frequent value)
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])

# For Blood Type (categorical) - Use mode
df['Blood Type'] = df['Blood Type'].fillna(df['Blood Type'].mode()[0])

# For Blood Pressure (categorical) - Use mode
df['Blood Pressure'] = df['Blood Pressure'].fillna(df['Blood Pressure'].mode()[0])

# For Diabetes (binary categorical) - Use mode
df['Diabetes'] = df['Diabetes'].fillna(df['Diabetes'].mode()[0])

# For Smoking (binary categorical) - Use mode
df['Smoking'] = df['Smoking'].fillna(df['Smoking'].mode()[0])
```

Mean vs Median:

Mean: Good for normally distributed data without outliers

Median: Better for skewed data or data with outliers (more robust)

Mode: Best for categorical data or discrete numerical data

🔹 Why Median for Age, Height, BMI, Heart Rate, Cholesterol:

These biological measurements often have outliers

Median is less affected by extreme values

More representative of typical values

🔹 Why Mean for Weight and Temperature:

These tend to have more normal distributions

Mean captures the average trend better

🔹 Why Mode for Categorical Variables:

Gender, Blood Type, Blood Pressure are categorical

Mode gives the most common category

Maintains the natural distribution of the data
